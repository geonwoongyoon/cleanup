```python
import pandas as pd
import pydeck as pdk

data = {
    'X': [126.9243, 126.9328, 126.9179, 126.9284],
    'Y': [37.52158, 37.52715, 37.52811, 37.51727],
    '이용합계': [3171175, 923672, 912157, 525694],
    '승차승객수': [1571550, 439038, 448979, 256975],
    '하차승객수': [1599625, 484634, 463178, 268719]
}


#이용합계 
#대여건수 승차
#반납건수 하차




df = pd.DataFrame(data)

# PyDeck HexagonLayer with '반납건수' as weight
layer = pdk.Layer(
    'HexagonLayer',
    data=df,
    get_position='[X, Y]',
    get_weight='하차승객수',  # Using '반납건수' as weight
    elevation_scale=10,
    radius=549,
    extruded=True,
    coverage=1,
    opacity=0.5,
    color_range=[[255, 255, 178], [254, 204, 92], [253, 141, 60]], 
    
   # get_fill_color='[[255, 0, 0, 150,], [255, 255, 178], [254, 204, 92], [253, 141, 60], [240, 59, 32]]',  # Red color with alpha
)
   #  color_range=[[255, 255, 178], [254, 204, 92], [253, 141, 60], [240, 59, 32]],# Set the viewport

# Set the viewport
view_state = pdk.ViewState(
    latitude=df['Y'].mean(),
    longitude=df['X'].mean(),
    zoom=10,
    pitch=15,
    bearing=-200
)

# Create a PyDeck deck
deck = pdk.Deck(layers=[layer], initial_view_state=view_state)

# Show the deck
deck.to_html('weighted_map_return_count_pydeck.html', iframe_width=800, iframe_height=500)
```
